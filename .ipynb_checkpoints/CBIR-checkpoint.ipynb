### SIFT

```python
import cbir
# 
dataset = cbir.Dataset().subset[0:dataSize]

# 
orb = cbir.descriptors.Orb()
voc = cbir.encoders.VocabularyTree(n_branches=4, depth=4, descriptor=orb)

# 
features = voc.extract_features(dataset)

# 
voc.fit(features)

# database
database = cbir.Database(dataset, encoder=voc)

# 
database.index()

# database
database.save()

# 
query = "118000.jpg"
scores = database.retrieve(query)
database.show_results(query, scores, figsize=(30, 10))
```

## CNN
### AlexNet as Descriptor

```python
import cbir
# 
dataset = cbir.Dataset().subset[0:dataSize]

# 
alexdesvoc = cbir.encoders.VocabularyTree(n_branches=3, depth=3, descriptor=cbir.descriptors.AlexNet())
alexdesvoc.learn(dataset)

# database
alexdesdb = cbir.Database(dataset, encoder=alexdesvoc)

# 
alexdesdb.index()

# database
alexdesdb.save()

# 
query = "126400.jpg"
scores = alexdesdb.retrieve(query)
alexdesdb.show_results(query, scores, figsize=(30, 10))
```

### AlexNet as Encoder

```python
import cbir
# 
dataset = cbir.Dataset().subset[0:dataSize]

# database
alexencdb = cbir.Database(dataset, encoder=cbir.encoders.AlexNet())

# 
alexencdb.index()

# database
alexencdb.save()

# 
query = "126400.jpg"
scores = alexencdb.retrieve(query)
alexencdb.show_results(query, scores, figsize=(30, 10))
```
